```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    'User-Agent': (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
        "(KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"
    )
}


# List to store extracted data
extracted_data = []

for i in all_names:

    for j in all_names[str(i)]:
        url = "https://www.astro.com/astro-databank/" + j.replace(' ', '_')
        print(f"Fetching: {url}")
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")

            name_row = soup.find('b', string='Name')
            name = name_row.find_next('td').get_text(strip=True).split('Gender:')[0].strip() if name_row else None

            # Birthname
            birthname_row = soup.find('td', class_="metadata-label", string='Birthname')
            birthname = birthname_row.find_next('td').get_text(strip=True) if birthname_row else None

            # Gender
            gender_row = soup.find('a', title="Help:Gender")
            gender = gender_row.parent.get_text(strip=True).replace('Gender:', '').strip() if gender_row else None

            #  Date of Birth
            date_of_birth = soup.find('td', string='born on')
            date = date_of_birth.find_next('td').get_text(strip=True) if date_of_birth else None

            # Place of Birth
            place_of_birth = soup.find('td', string='Place')
            birth_place = place_of_birth.find_next('td').get_text(strip=True) if place_of_birth else None

            #  Timezone
            timezone_row = soup.find('a', title="Help:Timezone")
            timezone = timezone_row.find_next('td').get_text(strip=True) if timezone_row else None

            source_data = soup.find('a',title="Help:DataSource")
            source = source_data.parent.find_next('td').get_text(strip=True).split('Rodden')[0].strip() if source_data else None
            #Rating
            Ratingg = soup.find('a', title="Help:RR")
            Rating = Ratingg.find_next('b').get_text(strip=True) if Ratingg else None

            Collectorr = soup.find('a',title ="Help:Collector")
            Collect = Collectorr.find_next('a').get_text(strip=True) if Collectorr else None

            # Astrology data
            Astrology_data = soup.find('a', title="Help:Astro data")
            a_data = Astrology_data.find_next('td').get_text(strip=True) if Astrology_data else None
            # print(a_data)
            sun, moon, asc = None, None, None

            if a_data:
                data_parts = a_data.split("Asc.")

                if len(data_parts) == 2:
                    sun_moon = data_parts[0]
                    asc = data_parts[1].strip().rstrip("'")

                    if "'" in sun_moon:
                        sun, moon = sun_moon.split("'")
                        sun = sun.strip()
                        moon = moon.strip()

                # carrier
                list_items = soup.find_all("li")
                # print(list_items)
                found_vocation = False
                for text in list_items:
                    teexxt = text.get_text()
                    if "Vocation" in teexxt:
                        Vocation = teexxt.split(':')
                        if len(Vocation) > 1:
                            career = Vocation[1]
                        found_vocation = True
                        break
                if not found_vocation:
                    career = None

                extracted_data.append({
                    'Name': name,
                    'Birthname': birthname,
                    'Gender': gender,
                    'Date of Birth': date,
                    'Place of Birth': birth_place,
                    'Timezone': timezone,
                    'Data Source' : source,
                    'Rodden Rating' : Rating,
                    'Collector' : Collect,
                    'Sun Position' : sun,
                    'Moon Position' : moon,
                    'Asc pos' : asc,
                    "Career " : career
                })

            else:
                print(f"Failed to fetch data for {j}: {response.status_code}")
```

```python
#48n33 48n0  48n3207 40n0116
# converting latitude to decimal
import re

def convert_to_decimal(lat):
    match = re.match(r"(\d+)[nNsS](\d{1,2})(\d{2})?", lat)

    if match:
        degrees = match.group(1)
        if match.group(3):
            minutes = match.group(2) + match.group(3)
        else:
            minutes = match.group(2)

        decimal_degrees = float(degrees) + float(minutes) / 60
        if 's' in lat.lower():
            decimal_degrees = -decimal_degrees
        return round(decimal_degrees, 2)
    else:
        return None

# 40n0116 48n33
# get latitude
df['latitude'] = df['get_latitude'].apply(convert_to_decimal)
```

```python
# Removing all rows where latitude is less than -90 and greater than 90
def lat(row):
  return -90 <= row <= 90
df = df[df['latitude'].apply(lat)]
```

```python
# convert longitude to decimal
def convert_to_decimal(long):
    try:
        match = re.match(r"(\d+)([eEwW])(\d{1,2})?(\d{2})?", long)
        if match:
            degrees = match.group(1)
            direction = match.group(2).lower()
            minutes = match.group(3) if match.group(3) else "0"
            if match.group(4):
                minutes += match.group(4)

            decimal_degrees = float(degrees) + float(minutes) / 60
            if direction == 'w':
                decimal_degrees = -decimal_degrees
            return round(decimal_degrees, 2)
        else:
            return None
    except ValueError:
        return None

# get longitude
df['longitude'] = df['get_longitude'].str.strip().apply(convert_to_decimal)
```

```python
# removing all rows where longitude value less than -180 and greater than 190
def longg(row):
  return -180 <= row <= 180
df = df[df['longitude'].apply(longg)]
```

```python
# Getting UTC time
import pandas as pd
from datetime import datetime
import pytz

def convert_to_utc(row):
    birth_datetime_str = f"{row['Birth Date']} {row['Birth Time']}"
    birth_datetime = datetime.strptime(birth_datetime_str, "%d %B %Y %H:%M")
    # print(birth_datetime)
    local_tz = pytz.timezone(row['UTC_Birth_Time_Zone'])
    # print(local_tz)
    local_time = local_tz.localize(birth_datetime)
    # Convert the local time to UTC
    utc_time = local_time.astimezone(pytz.utc)
    return utc_time.strftime("%H:%M")

df['UTC_Birth_Time'] = df.apply(convert_to_utc, axis=1)

# print(df1)
```

```python
import swisseph as swe
from datetime import datetime
import pandas as pd

def get_planet_positions(date_of_birth, time_utc):

    dob_datetime = datetime.strptime(date_of_birth, "%d %B %Y")
    year, month, day = dob_datetime.year, dob_datetime.month, dob_datetime.day

    hh, mm = map(int, time_utc.split(":"))
    hour = hh + mm / 60

    # Convert to Julian Day Number (JD)
    jd_ut = swe.julday(year, month, day, hour)

    # Define planets
    planets = {
        'Sun': swe.SUN, 'Moon': swe.MOON, 'Mercury': swe.MERCURY,
        'Venus': swe.VENUS, 'Mars': swe.MARS, 'Jupiter': swe.JUPITER,
        'Saturn': swe.SATURN, 'Uranus': swe.URANUS, 'Neptune': swe.NEPTUNE,
        'Pluto': swe.PLUTO, 'Rahu': swe.MEAN_NODE, 'Ketu': swe.TRUE_NODE
    }

    # Compute planetary positions
    planet_positions = {}
    for planet, code in planets.items():
        position, _ = swe.calc_ut(jd_ut, code)
        planet_positions[f'{planet}_degree'] = get_degree(position[0])
        planet_positions[f'{planet}_zodiac'] = get_zodiac_sign(position[0])

    return planet_positions

def get_degree(degrees):
    degree_in_sign = degrees % 30
    return f"{degree_in_sign:.2f}°"

def get_zodiac_sign(degrees):
    signs = ["Aries", "Taurus", "Gemini", "Cancer", "Leo", "Virgo",
             "Libra", "Scorpio", "Sagittarius", "Capricorn", "Aquarius", "Pisces"]

    sign_index = int(degrees // 30)
    return signs[sign_index]

planet_columns = df.apply(lambda row: pd.Series(get_planet_positions(row['Birth Date'], row['UTC_Birth_Time'])), axis=1)
df1 = pd.concat([df, planet_columns], axis=1)

print(df1)
```

```python
import datetime

def calculate_mahadasha(birth_date, birth_time):
    mahadashas = {
        'Sun': 6, 'Moon': 10, 'Mars': 7, 'Rahu': 18, 'Jupiter': 16, 'Saturn': 19,
        'Mercury': 17, 'Venus': 20, 'Ketu': 7
    }

    current_date = datetime.datetime.now()
    # print(current_date)
    birth_datetime = datetime.datetime.strptime(birth_date + " " + birth_time, "%d %B %Y %H:%M")
    # print(birth_datetime)
    age = current_date - birth_datetime
    # print(age)
    age_years = age.days // 365
    # print(age_years)
    # Determine the current Mahadasha
    current_mahadasha = None
    elapsed_years = 0
    for planet, duration in mahadashas.items():
        elapsed_years += duration
        if elapsed_years > age_years:
            current_mahadasha = planet
            break

    return current_mahadasha

df3['Dashas'] = df3.apply(lambda row: pd.Series(calculate_mahadasha(row['Birth Date'],row['UTC_Birth_Time'])), axis=1)


```
